```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x=np.arange(1, 31)
y=x*2

plt.plot(x, y)
plt.title('Linear Graph')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

```python
df=pd.read_csv('fdata.csv')
dates=df['Date'].tolist()
openData=df['Open'].tolist()
highData=df['High'].tolist()
lowData=df['Low'].tolist()
closeData=df['Close'].tolist()

print('Dates ', dates)
print('Open ', openData)
print('High ', highData)
print('Low ', lowData)
print('Close ', closeData)

plt.plot(dates, openData, color='r', label='Open', linewidth=2)
plt.plot(dates, highData, color='g', label='High', linewidth=3)
plt.plot(dates, lowData, color='b', label='Low', linewidth=4)
plt.plot(dates, closeData, color='c', label='Close', linewidth=5)

plt.legend()
plt.xlabel('Dates')
plt.ylabel('Stock Price')
```

```python
xval=[10, 20, 30]
y1val=[40, 10, 30]
y2val=[20, 40, 10]

plt.plot(xval, y2val, color='r', linewidth=3, linestyle='dotted', label='red-dotted')
plt.plot(xval, y1val, color='g', linewidth=5, linestyle='dashed', label='green-dashed')

plt.legend(loc=1)
plt.title('Plot with two or more lines with different styles')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
prog_lang=['Python', 'Java', 'C', 'C++', 'R', 'JavaScript', 'C#']
pop=[100, 96.3, 94.4, 87.5, 81.5, 79.4, 74.5]
plt.bar(prog_lang, pop)
plt.title('Popularity of programming languages 2019 according to IEEE')
```

```python
plt.barh(prog_lang, pop)
plt.title('Popularity of programming languages 2019 according to IEEE')
```

```python
bargp=plt.bar(prog_lang, pop)
colors=('r', 'g', 'b', 'c', 'y', 'm', 'k')
for i in range(7):
    bargp[i].set_color(colors[i])

plt.show()
```

```python
plt.bar(prog_lang, pop)
for i in range(len(prog_lang)):
    plt.text(prog_lang[i], pop[i]+2, pop[i])
```

```python
explodes = (0.2, 0, 0, 0, 0, 0, 0)
plt.pie(pop, labels=prog_lang, explode=explodes, shadow=True, startangle=90, autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```

```python
mean_velocity=[0.2474, 0.1235, 0.1737, 0.1824]
std_velocity=[ 0.3314, 0.2278, 0.2836, 0.2645]
plt.bar(np.arange(0,4), mean_velocity, yerr=std_velocity,
       error_kw={'ecolor': '0.1', 'capsize': 10}, alpha=0.7, label='velocity')
plt.legend()
```
