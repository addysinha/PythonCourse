```python
import csv
import random

name=input('Please enter your name ? ')
text=[name,]
score=0

file=open('questions.csv', 'r')

randomnum=[random.randint(0, 3), random.randint(0, 3)]
while randomnum[0] == randomnum[1]:
    randomnum=[random.randint(0, 3), random.randint(0, 3)]

linenum=0
questioncount=0
for i, line in csv.reader(file):
    if linenum in randomnum:
        questioncount+=1
        ans=input(i)
        text.append([i, line])
        if ans == line:
            score+=1
        if questioncount == 2:
            text.append(score)
    linenum += 1

print(text)

file.close()
```

```python
import csv
menu=0
while menu != 4:
    menu=int(input('''Menu:
    1. Add to file
    2. View all records
    3. Delete a record
    4. Quit Programme
    Please select your menu option ? '''))
    if menu == 1:
        try:
            file = open('Salaries.csv', 'a')
            name=input('Please enter your name: ')
            salary=int(input('Please enter your salary: '))
            file.write(name + ',' + str(salary) + '\n')
            file.close()
        except IOError:
            print('Exception: Unable to perform file operation')
    elif menu == 2:
        try:
            file=open('Salaries.csv', 'r')
            for line in list(csv.reader(file)):
                print(line)
            file.close()
        except FileNotFoundError:
            print('Exception: Unable to find the file Salaries.csv')
    elif menu == 3:
        try:
            file=open('Salaries.csv', 'r')
            tmp=[]
            name=input('Please provide the name of the person, for whom you want to remove the record ? ')
            for linelist in list(csv.reader(file)):
                if not name in linelist:
                    tmp.append(linelist)
            
            print(tmp)            
            file.close()
            
            file=open('Salaries.csv', 'w')
            for linelist in tmp:
                file.write(linelist[0] + ',' + linelist[1] + '\n')
            file.close()
        except FileNotFoundError:
            print('Exception: Unable to find the file Salaries.csv')
    elif not menu in (1, 2, 3, 4):
        print('Error: Please enter a valid menu')
```

```python
import re

total=0.0
values=0
while values != '':
    print('Please keep on entering the values below to finally get a sum: ')
    values=input('').strip()
    if re.match('\d.\d|\d|.\d', values):
        total+=float(values)
        print('Sum: ', total)
    elif values.strip() != '':
        print('Entered value is not a number. Please enter a number')
```

```python
from tkinter import *
import csv

def add():
    value=textbox1.get()
    if value.isdigit():
        listbox1.insert(END, value)
        textbox1.delete(0, END)
    else:
        textbox1.delete(0, END)

def clear():
    textbox1.delete(0, END)
    listbox1.delete(0, END)
    textbox1.focus()

def save():
    tmp_list=listbox1.get(0, END)
    try:
        file=open('ListName.csv', 'w')
        for name in tmp_list:
            file.write(name + ',')
    except IOError:
        print('Exception: Some issue with the file. Please retry.')
    file.close()
    
window=Tk()
window.title('Whole number List')
window.geometry('850x500')

label1=Label(text='Please enter a whole number ?')
label1.place(x=40, y=10)

textbox1=Entry('')
textbox1.place(x=225, y=10, width=100, height=30)
textbox1.focus()

button1=Button(text='Add', command=add)
button1.place(x=350, y=10, width=50, height=30)

button2=Button(text='Clear', command=clear)
button2.place(x=425, y=10, width=50, height=30)

button3=Button(text='Save', command=save)
button3.place(x=500, y=10, width=50, height=30)

listbox1=Listbox()
listbox1.place(x=150, y=50, width=50, height=100)

window.mainloop()
```

```python
from tkinter import *
import re
import csv

def create():
    filename=filenametextbox.get().strip()
    if re.match('\w+\.csv', filename):
        try:
            file=open(filename, 'w')
            file.close()
            filenamelabel2['text']=filename + ' has been created'
        except IOError:
            filenamelabel2['text']='Exception: IOError occurred'
    else:
        try:
            file=open(filename + '.csv', 'w')
            file.close()
            filenamelabel2['text']=filename + '.csv has been created'
        except IOError:
            filenamelabel2['text']='Exception: IOError occurred'

def save():
    filename=filenametextbox.get().strip()
    if re.match('\w+\.csv', filename):
        try:
            file=open(filename, 'a')
            name=textbox1.get().strip()
            age=textbox2.get().strip()
            file.write(name + ',' + age + '\n')
            file.close()
        except FileNotFoundError:
            filenamelabel2['text']='Exception: File Not Found'
    else:
        try:
            file=open(filename + '.csv', 'a')
            name=textbox1.get().strip()
            age=textbox2.get().strip()
            file.write(name + ',' + age + '\n')
            file.close()
        except FileNotFoundError:
            filenamelabel2['text']='Exception: File Not Found'

def extract():
    filename=filenametextbox.get().strip()
    listbox1.delete(0, END)
    if re.match('\w+\.csv', filename):
        try:
            file=open(filename, 'r')
            for name, age in list(csv.reader(file)):
                listbox1.insert(END, name + ',' + age)
            file.close()
        except FileNotFoundError:
            filenamelabel2['text']='Exception: File Not Found'
    else:
        try:
            file=open(filename + '.csv', 'r')
            for name, age in list(csv.reader(file)):
                listbox1.insert(END, name + ',' + age)
            file.close()
        except FileNotFoundError:
            filenamelabel2['text']='Exception: File Not Found'

window=Tk()
window.title('Data Entry Screen')
window.geometry('750x500')

filenamelabel1=Label(text='Please enter the file name to create ?')
filenamelabel1.place(x=50, y=10)

filenametextbox=Entry('')
filenametextbox.place(x=250, y=10, width=100, height=30)

filenamebutton=Button(text='Create the file', command=create)
filenamebutton.place(x=375, y=10, width=100, height=30)

filenamelabel2=Label(text='')
filenamelabel2.place(x=500, y=10)

label1=Label(text='Please enter the name ? ')
label1.place(x=50, y=50)

textbox1=Entry('')
textbox1.place(x=200, y=50, width=100, height=30)

label2=Label(text='Please enter the age ? ')
label2.place(x=50, y=100)

textbox2=Entry('')
textbox2.place(x=200, y=100, width=100, height=30)

button1=Button(text='Save', command=save)
button1.place(x=75, y=150, width=50, height=30)

button2=Button(text='Extract', command=extract)
button2.place(x=50, y=250, width=50, height=30)

listbox1=Listbox()
listbox1.place(x=50, y=300, width=50, height=100)

window.mainloop()
```
